```python
df_New_Cases = pd.read_csv('New Cases - vaccination.csv')
df_New_Death = pd.read_csv('New Death - vaccination.csv')
df_Severe_Cases = pd.read_csv('Severe Cases - vaccination.csv')
df_Culmulativ = pd.read_csv('Culmulativ pecentage of vaccines.csv')

```

```python
df_New_Cases.info()
```

```python
df_New_Cases['Date'] = pd.to_datetime(df_New_Cases['Date'],format = '%d-%m-%Y')
df_New_Death['Date'] = pd.to_datetime(df_New_Death['Date'],format = '%d-%m-%Y')
df_Severe_Cases['Date'] = pd.to_datetime(df_Severe_Cases['Date'],format = '%d-%m-%Y')
df_Culmulativ['Date'] = pd.to_datetime(df_Culmulativ['Date'],format = '%d-%m-%Y')
```

```python
df_New_Cases.head(6) 
```

```python
df = df_New_Cases.merge(df_New_Death,on = ['Date','age group']) 
```

```python
df1 = df.merge(df_Severe_Cases,on = ['Date','age group']) 
```

```python
df1
```

```python
df1.columns
```

```python
df1.drop(['New cases vaccinated per 100k','New cases expired Vaccination  per 100k','New cases Not vaccinated per 100k','New Death  vaccinated per 100k','New Death  expired Vaccination  per 100k','New Death  Not vaccinated per 100k','Severe cases expired Vaccination  per 100k','New severe cases vaccinated per 100k','New severe cases expired Vaccination  per 100k','New severe cases Not vaccinated per 100k','Severe cases vaccinated per 100k','Severe cases Not vaccinated per 100k'],axis = 1,inplace=True)
```

```python
df1.info()
```

```python
now = datetime.now()

current_time = now.strftime("%d_%m_%Y_%H_%M")
print("Current Time =", current_time)

filename = 'corona'+ current_time +'.csv'
with open(filename,mode='w',encoding='UTF-16',newline='') as rawdata_file:
    header = df1.columns
    writer = csv.writer(rawdata_file)
    writer.writerow(header)
    
    for row,row_data in df1.iterrows():
        writer.writerow(row_data)
```

```python
allpop = df1[df1['age group'] == 'All population']
```

```python
now = datetime.now()

current_time = now.strftime("%d_%m_%Y")
print("Current Time =", current_time)

filename = 'allpopcorona'+ current_time +'.csv'
with open(filename,mode='w',encoding='UTF-16',newline='') as rawdata_file:
    header = allpop.columns
    writer = csv.writer(rawdata_file)
    writer.writerow(header)
    
    for row,row_data in allpop.iterrows():
        writer.writerow(row_data)
```

```python
notallpop = df1[df1['age group'] != 'All population']
```

```python
now = datetime.now()

current_time = now.strftime("%d_%m_%Y")
print("Current Time =", current_time)

filename = 'notallpopcorona'+ current_time +'.csv'
with open(filename,mode='w',encoding='UTF-16',newline='') as rawdata_file:
    header = notallpop.columns
    writer = csv.writer(rawdata_file)
    writer.writerow(header)
    
    for row,row_data in notallpop.iterrows():
        writer.writerow(row_data)
```
